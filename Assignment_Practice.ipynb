```python
df = pd.read_csv("/content/carsales.csv")
df
```

```python
year = df.groupby('Year')['Price'].median()
year
```

```python
plt.scatter(df['Price'], df['Milage'], marker = 8, linewidths = 10)
plt.title("Scatter Plot")
plt.xlabel("Price")
plt.ylabel("Mileage")
plt.grid(True)
plt.show()
```
