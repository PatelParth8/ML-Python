```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

SET : 1

```python
# Load the CSV file named "carsales.csv" into a Pandas DataFrame.

df = pd.read_csv("carsales.csv")
df
```

```python
# Display the first 5 rows.

print(df.head(5))
```

```python
# Calculate the mode of the "Make" column using Pandas.

print(df.mode())
print(df['Make'].value_counts())
```

```python
# Group the DataFrame by the "Year" column and calculate the median price for each year.

year = df.groupby('Year')['Price'].median()
year
```

```python
# Create a scatter plot using Matplotlib to visualize the relationship between the "Price"
# and "Mileage" columns.

plt.scatter(df['Price'], df['Milage'], marker=8, linewidths=15)
plt.title("Scatter Plot")
plt.xlabel("Mileage")
plt.ylabel("Price")
plt.grid(True)
plt.show()
```

```python
# Filter the DataFrame to only include rows where the "FuelType" column is 'Petrol'.
# Display the first 5 rows.

print(df[df['FuelType'] == 'Petrol'].head(5))
```

```python
# Calculate the total number of missing values in each column of the DataFrame using
# Pandas.

print(df.isnull().sum())
```

```python
# Create a bar chart using Matplotlib to show the count of each unique value in the
# "BodyType" column.

count = df['BodyType'].value_counts()
count
```

```python
plt.bar(count.index, count)
plt.title("Bar Chart")
plt.xlabel("Car Name")
plt.ylabel("Index Value")
plt.grid(True)
plt.show()
```

```python
# Calculate the standard deviation of the "EngineSize" column using NumPy.

#df['EngineSize'] = df['EngineSize'].str.replace(r'\D', '').astype(int)
n = np.std(df['EngineSize'])
n
```

```python
# Plot a histogram of the "Price" column using Pandas.

plt.hist(df['Price'])
plt.title("Histogram")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
# Filter the DataFrame to only include rows where the "Year" is greater than 2015. Display
# the last 5 rows.

print(df[df['Year'] > 2015].tail(5))
```
