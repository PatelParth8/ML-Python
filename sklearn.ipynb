```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# **LINEAR REGRESSION**

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('iris')
df
```

```python
df['species'].value_counts()
```

```python
# Filter Independent Variable

x = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
x
```

```python
x.shape
```

```python
# Filter Dependent Variable

y = df['species']
y
```

```python
y.shape
```

```python
# Conver Y in numbers Encoding

from sklearn.preprocessing import LabelEncoder
```

```python
le_y = LabelEncoder()
y = le_y.fit_transform(y)
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
x_train
```

```python
x_train.size
```

```python
x_test.size
```

```python
y_train.size
```

```python
y_test.size
```

# **LOGISTIC REGRESSION**

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
log.fit(x_train, y_train)
```

```python
y_pred = log.predict(x_test)
y_pred
np.sort(y_pred)
```

```python
y_test
np.sort(y_test)
```

# **CONFUSION MATRIX**

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, y_pred)
```

```python
(28/30)*100
```

```python
log.score(x_test, y_test)
```

# **KNEIGHBOURS CLASSIFIERS**

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
classifier_knn= KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2) # distance name
```

```python
classifier_knn.fit(x_train, y_train)
```

```python
y_pred_kn = classifier_knn.predict(x_test)
y_pred_kn
```

```python
# confusion matrix
confusion_matrix(y_test, y_pred_kn)
```

```python
classifier_knn.score(x_test, y_test)
```

# **ACCURACY SCORE**

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred_kn, y_test)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
classifier_nb = GaussianNB()
```
