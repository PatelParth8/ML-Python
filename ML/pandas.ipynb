# **Panda Series**

```python
list = [1,2,-3,6.2, 'data values']
print(list)
```

```python
series1 = pd.Series(list)
print(series1)
```

```python
type(series1)
```

```python
series2 = pd.Series([1,2,3,4])
print(series2)
```

```python
a = pd.Series([])
print(a)
```

```python
series3 = pd.Series([1,2,3,4], index = ['a', 'b', 'c', 'd'], dtype = float)
print(series3)
```

```python
series3 = pd.Series([1,2,3,4], index = ['a', 'b', 'c', 'd'], dtype = float, name = 'data values')
print(series3)
```

```python
scaler = pd.Series(0.5)
print(scaler)
```

```python
scaler = pd.Series(0.5, index = [1,2,3])
print(scaler)
```

```python
dic = pd.Series({'a': 1, 'b': 2})
print(dic)
```

```python
s = pd.Series([1,2,3,4,5])
print(s)
```

```python
#indexing
s[0]
```

```python
#indexing
s[3]
```

```python
#slicing
s[0:3]
```

```python
#maximum
max(s)
```

```python
#minimum
min(s)
```

```python
#conditional operator
s[s > 3]
```

```python
s1 = pd.Series([1,2,3,4,5])
print(s1)
```

```python
#addition
s + s1
```
