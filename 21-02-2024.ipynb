```python
#count
car_make = df['Car Make'].value_counts()
car_make
```

```python
car_model = df['Car Model'].value_counts()
car_model
```

```python
#Pie Chart
fig, ax = plt.subplots(figsize=(4, 4))
ax.pie(car_make, labels=car_make.index, autopct='%1.1f%%', startangle=90)
ax.axis('equal')
ax.set_title('Distribution of Car Makes')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(4, 4))
ax.pie(car_model, labels=car_model.index, autopct='%1.1f%%')
ax.axis('equal')
ax.set_title('Distribution of Car Models')
plt.show()
```
