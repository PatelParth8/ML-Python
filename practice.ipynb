```python
plt.barh(df['Name'], df['Gross Salary'])
plt.title("Horizontal Bar Chart")
plt.xlabel("Gross Salary")
plt.ylabel("Name")
plt.grid(axis = 'x')
plt.show()
```

```python
plt.scatter(df['Name'], df['Salary'], linewidth = 10, marker = 8)
plt.title("Scatter Plot")
plt.xlabel("Name")
plt.ylabel("Salary")
plt.grid(True)
plt.show()
```

```python
plt.hist(df['Gross Salary'], bins = 10, edgecolor = 'black')
plt.title("Histogram")
plt.xlabel("Gross Salary")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
data = {'A': range(1,11), 'B': [5,7,3,8,10,6,2,4,9,3]}
data

df = pd.DataFrame(data)
df
```

```python
#Line Chart
plt.plot(df['A'], df['B'], marker = "o", linestyle = "dashed")
plt.title("Line Chart")
plt.xlabel("A")
plt.ylabel("B")
plt.grid(True)
plt.show()
```

```python
data = {'A': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'], 'B': [5,7,3,8,10,6,2,4,9,3]}
data

df = pd.DataFrame(data)
df
```

```python
#Line Chart
plt.plot(df['A'], df['B'], marker = "o", linestyle = 'dashed')
plt.title("Line Chart")
plt.xlabel("A")
plt.ylabel("B")
plt.grid(True)
plt.show()
```

```python
plt.bar(df['A'], df['B'])
plt.title("Bar Chart")
plt.xlabel("A")
plt.ylabel("B")
plt.grid(axis = 'y')
plt.show()
```

```python
plt.scatter(df['A'], df['B'], linewidths = 5, marker = 8)
plt.title("Scatter Plot")
plt.xlabel("A")
plt.ylabel("B")
plt.grid(True)
plt.show()
```

```python
plt.pie(df['B'], labels=df['A'], autopct="%1.1f%%", startangle=90)
plt.title("Pie Chart")
plt.show()
```

```python
plt.hist(df['B'], bins = 10, edgecolor = 'black')
plt.title("Histogram")
plt.xlabel("B")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
