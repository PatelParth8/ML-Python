**SECTION : B**

```python
# Load the fracture.csv data into pandas dataframe

df = pd.read_csv("/content/fracture.csv")
df
```

```python
# print the first 15 record

print(df.head(15))
```

```python
# Add a new column named “bmi” to store Body Mass Index for each record. Formula to calculate BMI is: weight_kg / (height_cm/100)^2.

df['bmi'] = df['weight_kg']/((df['height_cm']/100) * (df['height_cm']/100))
df
```

```python
# Split the data set into test and train.

x = df[['age', 'bmd', 'bmi']]
x
```

```python
y = df[['sex']]
y
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
y = le.fit_transform(y)
y
```

```python
# Split the data set into test and train.

from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
x_train
```

```python
# Build a logistic regression model to predict the fracture considering age, sex, bmi and bmd.

from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred
```

```python
lr.score(x_test, y_test)
```

```python
# Calculate accuracy of the model using a confusion matrix.

from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred, y_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, y_pred)
```

```python
# Apply the model on to the test and train data. Plot the outcomes of prediction using appropriate graphs.

plt.scatter(y_test, y_pred)
plt.title("y-test & y-prediction")
plt.xlabel("Y-Test")
plt.ylabel("Y-Prediction")
plt.show()
```
