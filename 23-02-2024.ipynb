```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {'a': range(1,11), 'b': [5,7,3,8,10,6,2,4,9,3]}
data
```

```python
df = pd.DataFrame(data)
df
```

```python
#LineChart
plt.plot(df['a'], df['b'], marker = 'o', linestyle = 'dashed')
plt.title('Line Chart')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid(True)
plt.show()
```

```python
data1 = {'A': ['A', 'B', 'C', 'D', 'E'], 'B': [15, 30, 25, 10, 20]}
data1

df1 = pd.DataFrame(data1)
df1
```

```python
#Vertical Bar Chart
plt.bar(df1['A'], df1['B'])
plt.title('Vertical Bar Chart')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid(axis='y')
plt.show()
```

```python
#Horizontal Bar Chart
plt.barh(df1['A'], df1['B'])
plt.title('Horizontal Bar Chart')
plt.xlabel('x-axis')
plt.ylabel('y-axias')
plt.grid(axis='x')
plt.show()
```

```python
#Scatter Plot
plt.scatter(df['a'], df['b'], linewidths=15, marker=8)
plt.title('Scatter Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.show()
```

```python
plt.pie(df1['B'], labels=df1['A'], autopct='%1.1f%%', startangle=90)
plt.title("Pie Chart")
plt.show()
```

```python
data2 = {'Values': np.random.randn(1000)}
data2

df2 = pd.DataFrame(data2)
df2
```

```python
#Histogram
plt.hist(df2['Values'], bins=10, edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
