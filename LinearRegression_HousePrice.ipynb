```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/student/Downloads/housing.csv')
df
```

```python
df.info()
```

```python
# We'll split the data into features (X) and target variable (Y).
X = df.drop('Price', axis=1)
X

Y = df['Price']
Y

# x = df.iloc[:-1]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# split data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
# random_state is optional

# initialize and train the model
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
# Make Predictions
Y_pred = model.predict(X_test)
print(Y_pred)
```

```python
Y_pred = model.predict([[5000, 10, 5, 1]])
print(Y_pred)
```

```python
print("Actual Prices :",Y_test)
print("Predicted Prices :",Y_pred)
```

```python
df1 = pd.DataFrame({'Actual Price': Y_test, 'Predicted Price': Y_pred})
df1
```

```python
# SCATTER PLOT
plt.scatter(Y_test, Y_pred)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual Vs. Predicted Prices')
plt.show()
```
